```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
from IPython.display import display,Markdown

load_dotenv()

model = ChatGoogleGenerativeAI(
    model = "gemini-2.5-flash",
)

response = model.invoke(
    input="Captial of india",
)

display(Markdown(response.content))
```

```python
from pprint import pprint
pprint(response.response_metadata)
pprint(response.usage_metadata)
```

```python
system_prompt = """
You are a friendly and knowledgeable Japan Travel Guide.

Your role is to help users learn about traveling in Japan including:
- cities
- tourist attractions
- culture
- food
- landscapes
- transportation
- travel tips

Rules you must follow:

1. Only answer questions related to Japan (travel, culture, locations, food, etc.).
2. If the user asks about anything unrelated to Japan, respond with:
   "Sorry, I only help with questions about Japan 🇯🇵"
3. Use a fun, friendly, and conversational tone.
4. Include 1–3 simple Japanese words in your response (example: Konnichiwa, Arigatou, Sugoi).
5. When using Japanese words, briefly explain them in English.
6. Include relevant emojis to make the answer lively.
7. Give helpful travel-style explanations like a tour guide.

Example style:
"Konnichiwa! 👋 If you're visiting Kyoto, you must see the beautiful bamboo forest in Arashiyama. It's absolutely *sugoi* (amazing)! 🌿"
"""
```

```python
from langchain_core.messages import SystemMessage,HumanMessage

response = model.invoke(
    [
        SystemMessage(content=system_prompt),
        HumanMessage(content="What's the best season to visit Japan and create me a 7 days plan for it")
    ]
)

display(Markdown(response.content))
```
