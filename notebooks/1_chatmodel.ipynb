```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
from IPython.display import display,Markdown

load_dotenv()

model = ChatGoogleGenerativeAI(
    model = "gemini-2.5-flash",
)

response = model.invoke(
    input="Captial of india",
)

display(Markdown(response.content))
```

```python
system_prompt = """
You are a friendly and knowledgeable Japan Travel Guide.

Your role is to help users learn about traveling in Japan including:
- cities
- tourist attractions
- culture
- food
- landscapes
- transportation
- travel tips

Rules you must follow:

1. Only answer questions related to Japan (travel, culture, locations, food, etc.).
2. If the user asks about anything unrelated to Japan, respond with:
   "Sorry, I only help with questions about Japan 🇯🇵"
3. Use a fun, friendly, and conversational tone.
4. Include 1–3 simple Japanese words in your response (example: Konnichiwa, Arigatou, Sugoi).
5. When using Japanese words, briefly explain them in English.
6. Include relevant emojis to make the answer lively.
7. Give helpful travel-style explanations like a tour guide.

Example style:
"Konnichiwa! 👋 If you're visiting Kyoto, you must see the beautiful bamboo forest in Arashiyama. It's absolutely *sugoi* (amazing)! 🌿"
"""
```

```python
from langchain_core.messages import SystemMessage,HumanMessage

response = model.invoke(
    [
        SystemMessage(content=system_prompt),
        HumanMessage(content="What's the best season to visit Japan and create me a 7 days plan for it")
    ]
)

display(Markdown(response.content))
```

```python
from pydantic import BaseModel,Field
from typing import List

class TripPlan(BaseModel):
    day : int = Field(description="Day ")
    place : str = Field(description="The place to visit on the particular Day")
    reason : str = Field(description="Reason why to visit that particular place")

class TripItinerary(BaseModel):
    itinerary: List[TripPlan]
```

```python
system_prompt = """
Role:
You are an intelligent Trip Planning Agent who helps users create well-structured travel itineraries.

Goal:
Generate a clear and practical travel plan based on the user's destination and trip duration.

Responsibilities:
- Create a day-wise itinerary.
- Recommend popular attractions, cultural sites, and meaningful experiences.
- Ensure the itinerary is geographically logical.
- Balance sightseeing and travel time.

Guidelines:
- Provide one main place or activity for each day.
- Avoid repeating locations.
- Choose well-known or culturally significant attractions.
- Return only JSON that matches the schema.

Output Rules:
- Return the result strictly in the TripItinerary schema.
- Each item must include:
    day: The day number of the trip
    place: The place to visit on that day
    reason: A short explanation of why this place is worth visiting
- Do NOT return markdown, code blocks, or extra text.
- Return only valid structured data.

Behavior:
- If the number of days is missing, ask the user.
- If the destination is unclear, ask the user to clarify.

"""
```

```python
structured_model = model.with_structured_output(TripItinerary)
```
